```python
import json
import matplotlib.pyplot as plt
import numpy as np

from kernel.kernel import Kernel

gen_files = ['100_2_025_005/', '100_3_03_01/', '200_2_03_005/', 'unbalanced/']

algorithms = ['KMeans_', 'Ward_']

kernels = ['PlainWalk', 'LogPlainWalk', 'Communicability', 'LogCommunicability',
           'Forest', 'LogForest', 'Heat', 'LogHeat', 'SigmoidCommuteTime', 
           'LogSigmoidCommuteTime']

kmeans_results = {}
ward_results = {}


for algorithm in algorithms:
    for kernel in kernels: 
        with open('src/results/my/generated_graphs/sklearn/'
                  + gen_files[3] + algorithm + kernel + '.txt', 'r') as file:
            results = json.loads(file.read())
            res = results.pop(kernel)
            if algorithm == 'KMeans_':
                kmeans_results[kernel] = res
            else:
                ward_results[kernel] = res
```

```python
plt.figure(figsize=(10, 10))
plt.plot(Kernel.default_params, kmeans_results[kernels[0]], 
                     label=algorithms[0]+kernels[0])
plt.plot(Kernel.default_params, kmeans_results[kernels[1]], 
                     label=algorithms[0]+kernels[1])
plt.plot(Kernel.default_params, ward_results[kernels[0]], 
                     label=algorithms[1]+kernels[0])
plt.plot(Kernel.default_params, ward_results[kernels[1]], 
                     label=algorithms[1]+kernels[1])
# plt.ylim(0.8)
# plt.title(gen_files[0])
# plt.legend()
# plt.show()

plt.plot(Kernel.default_params, kmeans_results[kernels[2]], 
                     label=algorithms[0]+kernels[2])
plt.plot(Kernel.default_params, kmeans_results[kernels[3]], 
                     label=algorithms[0]+kernels[3])
plt.plot(Kernel.default_params, ward_results[kernels[2]], 
                     label=algorithms[1]+kernels[2])
plt.plot(Kernel.default_params, ward_results[kernels[3]], 
                     label=algorithms[1]+kernels[3])
# plt.ylim(0.7)
# plt.xlim(0.05, 0.3)
plt.title(gen_files[0])
plt.xlabel('Парметр')
plt.ylabel('ARI')
# plt.legend(bbox_to_anchor=(1., 1), loc=2, borderaxespad=0.5)
plt.legend()
plt.savefig('src/ploot.png')
# plt.show()

plt.plot(Kernel.default_params, kmeans_results[kernels[4]], 
                     label=algorithms[0]+kernels[4])
plt.plot(Kernel.default_params, kmeans_results[kernels[5]], 
                     label=algorithms[0]+kernels[5])
plt.plot(Kernel.default_params, ward_results[kernels[4]], 
                     label=algorithms[1]+kernels[4])
plt.plot(Kernel.default_params, ward_results[kernels[5]], 
                     label=algorithms[1]+kernels[5])
# plt.ylim(0.7)
# plt.xlim(0.05, 0.3)
plt.title(gen_files[0])
plt.legend()
plt.show()

plt.plot(Kernel.default_params, kmeans_results[kernels[6]], 
                     label=algorithms[0]+kernels[6])
plt.plot(Kernel.default_params, kmeans_results[kernels[7]], 
                     label=algorithms[0]+kernels[7])
plt.plot(Kernel.default_params, ward_results[kernels[6]], 
                     label=algorithms[1]+kernels[6])
plt.plot(Kernel.default_params, ward_results[kernels[7]], 
                     label=algorithms[1]+kernels[7])
plt.ylim(0.7)
plt.xlim(0.05, 0.9)
plt.title(gen_files[0])
plt.legend()
plt.show()

plt.plot(Kernel.default_params, kmeans_results[kernels[8]], 
                     label=algorithms[0]+kernels[8])
plt.plot(Kernel.default_params, kmeans_results[kernels[9]], 
                     label=algorithms[0]+kernels[9])
plt.plot(Kernel.default_params, ward_results[kernels[8]], 
                     label=algorithms[1]+kernels[8])
plt.plot(Kernel.default_params, ward_results[kernels[9]], 
                     label=algorithms[1]+kernels[9])
# plt.ylim(0.7)
# plt.xlim(0.05, 0.9)
plt.title(gen_files[0])
plt.legend()
plt.show()
```
